```python
n = int(input())
arr = list(map(int, input().split()))

freq = {}          
for x in arr:
    if x in freq:
        freq[x] += 1
    else:
        freq[x] = 1

delete = 0
for x in freq:
    f = freq[x]
    if f > x:
        delete += f - x
    elif f < x:
        delete += f

print(delete)
```

```python
s = input().strip()
k = int(input())

max_num = s[0:k]

for i in range(1, len(s) - k + 1):
    sub = s[i:i+k]
    if sub > max_num:
        max_num = sub

print(max_num)
```

```python
def hop_le(s):
    stack = []
    cap = {')': '(', '}': '{', ']': '['}

    for c in s:
        if c in "({[":
            stack.append(c)
        else:
            if not stack:
                return False
            if stack.pop() != cap[c]:
                return False

    return len(stack) == 0

s = input()
if hop_le(s):
    print("Hợp lệ")
else:
    print("Không hợp lệ")
```

```python
def daytang(nums):
    n= len(nums)
    dp= [1]*n
    for i in range(n):
        for j in range(i):
            if nums[j] < nums[i]:
                dp[i]=max(dp[i], dp[j]+1)
    return max(dp)
nums =[1,2,2,5,4,78,9,6,3,21,5,5,8]
print(daytang(nums))
```
