```python
def solon(a, b) -> int:
    if a>b:
        return a
    elif b > a:
        return b
a, b = map(int, input().split())
print(solon(a, b))
```

```python
def swap(a, b) -> tuple[int, int]:
    a=a^b
    b=a^b
    a=a^b
    return a, b
a, b = map(int, input().split())
print(swap(a, b))
```

```python
def so_nguyen_to(n):
    if n<2:
        return False
    for i in range(2, int(n**0.5)+1):
        if n%i==0:
            return False
    return True
n = float(input())
print(so_nguyen_to(n))
```

```python
def so_hh(n)-> bool:
    tong=0
    for i in range(1, n):
        if n%i==0:
            tong+=i
    return tong==n
n = int(input())
print(so_hh(n))
    
```

```python
def gt(n)->int:
    giai_thua=1
    for i in range (1, n+1):
        giai_thua*=i
    return giai_thua
n = int(input())
print(gt(n))
```

```python
def vi_tri(list, k):
    for i in range(len(list)):
        if list[i] == k:
            return i
    return -1
list = input()
k= input()
print(vi_tri(list, k))
        
    
```

```python
def tong(n):
    tong=0
    while n>0:
        tong+= n%10
        n = n//10
    return tong 
n = int(input())
print(tong(n))
```

```python
def may_tinh(num1, num2, operat):
    if operat == '+':
        kq = num1 + num2
    elif operat == '-':
        kq = num1 - num2
    elif operat == '*':
        kq = num1 * num2
    elif operat == '/':
        if num2 == 0:
            return "Khong the chia cho 0"
        kq = num1 / num2
    else:
        return "Toan tu khong hop le"
    
    return round(kq, 2)
num1 = float(input())
num2 = float(input())
operat = input()

print(may_tinh(num1, num2, operat))

```

```python
def hamming_distance(x, y):
    xor = x ^ y      
    count = 0
    while xor > 0:
        count += xor & 1
        xor >>= 1
    return count
x, y = map(int, input().split())
print(hamming_distance(x, y))
```
