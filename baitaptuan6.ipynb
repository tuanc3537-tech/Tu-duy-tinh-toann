```python
lst = list(map(int, input().split()))

result = {
    'positives': 0,
    'negatives': 0,
    'zeros': 0
}

for x in lst:
    if x > 0:
        result['positives'] += 1
    elif x < 0:
        result['negatives'] += 1
    else:
        result['zeros'] += 1

print(result)
```

```python
d1 = {}
d2 = {}

pairs1 = input().split()
pairs2 = input().split()

for p in pairs1:
    k, v = p.split(":")
    d1[k] = int(v)

for p in pairs2:
    k, v = p.split(":")
    d2[k] = int(v)

result = d1.copy()

for k, v in d2.items():
    result[k] = result.get(k, 0) + v

result = dict(sorted(result.items()))
print(result)
```

```python
nums = list(map(int, input("Nhập danh sách số nguyên: ").split()))
k = int(input("Nhập k: "))

a = []

for i in range(len(nums)):
    for j in range(i + 1, len(nums)):
        if nums[i] + nums[j] == k:
            a.append((nums[i], nums[j]))
a.sort(key=lambda x: (x[0], x[1]))

print(a)
```

```python
lst  = list(map(int, input().split()))
a = {}

for i in lst:
    if i not in a:
        a[i] = 1
    else:
        a[i] += 1

m = max(a.values())

b = (num for num, c in a.items() if c == m)

print(min(b))
```
