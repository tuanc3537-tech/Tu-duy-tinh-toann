```python
def tim(lst, k):
    for i in range(len(lst)):
        if lst[i] == k:
            return i, lst[i]
    return -1
k = int(input())
lst = list(map(int, input().split()))
print(tim(lst, k))
```

```python
def anh_xa_chuoi_vi_tri():
    danh_sach_chuoi = input().split()
    ket_qua = {}
    for vi_tri in range(len(danh_sach_chuoi)):
        chuoi = danh_sach_chuoi[vi_tri]
        if chuoi in ket_qua:
            ket_qua[chuoi] += (vi_tri,)
        else:
            ket_qua[chuoi] = (vi_tri,)
    return ket_qua
print(anh_xa_chuoi_vi_tri())
```

```python
def so_lon_nhat(lst):
    sln = lst[0]          
    for i in range(1, len(lst)):
        if lst[i] > sln:
            sln = lst[i]
    return sln

lst = list(map(int, input().split()))
print(so_lon_nhat(lst))
```

```python
def remove_less_than(arr, x):
    i = 0
    while i < len(arr):
        if arr[i] < x:
            arr.pop(i)
        else:
            i += 1

lst = list(map(int, input().split()))
x = int(input("Nhập x: "))
remove_less_than(lst, x)
print(lst)
```

```python
def day_0_ve_cuoi(a):
    pos = 0
    for i in range(len(a)):
        if a[i] != 0:
            a[pos] = a[i]
            pos += 1
    while pos < len(a):
        a[pos] = 0
        pos += 1


a = list(map(int, input("Nhập mảng: ").split()))
day_0_ve_cuoi(a)
print("Sau khi đẩy 0 về cuối:", a)
```

```python
def day_0_ve_dau(a):
    pos = len(a) - 1
    for i in range(len(a) - 1, -1, -1):
        if a[i] != 0:
            a[pos] = a[i]
            pos -= 1
    while pos >= 0:
        a[pos] = 0
        pos -= 1


a = list(map(int, input("Nhập mảng: ").split()))
day_0_ve_dau(a)
print("Sau khi đẩy 0 về đầu:", a)
```

```python
def tim_so_thieu(a):
    khong_co = []
    for x in range(10):     
        if x not in a:      
            khong_co.append(x)
    return khong_co


a = list(map(int, input("Nhập mảng: ").split()))
kq = tim_so_thieu(a)

if len(kq) == 0:
    print("Mảng đã có đủ các số từ 0 đến 9")
else:
    print("Các số không xuất hiện trong mảng:", kq)
```

```python
def tong_cuc_dai_dia_phuong(a):
    tong = 0
    for i in range(1, len(a) - 1):   
        if a[i] > a[i-1] and a[i] > a[i+1]:
            tong += a[i]
    return tong


a = list(map(int, input("Nhập danh sách: ").split()))
print("Tổng các phần tử cực đại địa phương là:", tong_cuc_dai_dia_phuong(a))
```

```python
def tim_vi_tri_nho_hon_x(a, x):
    for i in range(len(a)):
        if a[i] < x:
            print(i, end=" ")   

a = list(map(int, input("Nhập dãy số: ").split()))
x = int(input("Nhập x: "))
print("Các vị trí có giá trị < x là:")
tim_vi_tri_nho_hon_x(a, x)
```

```python
def kiem_tra_day(a):
    tang = True
    giam = True

    for i in range(len(a)-1):
        if a[i] >= a[i+1]:
            tang = False
        if a[i] <= a[i+1]:
            giam = False

    if tang:
        return "Dãy tăng thực sự"
    elif giam:
        return "Dãy giảm thực sự"
    else:
        return "Dãy vô danh"
a = list(map(int, input("Nhập dãy số: ").split()))
print(kiem_tra_day(a))
```
