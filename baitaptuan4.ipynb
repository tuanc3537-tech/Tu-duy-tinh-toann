```python
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):   # kiểm tra ước
        if n % i == 0:
            return False
    return True
```

```python
n = int(input())
uoc_max = 1

i = 2
while i * i <= n:
    while n % i == 0:
        uoc_max = i      
        n //= i          
    i += 1

if n > 1:
    uoc_max = n

print(uoc_max)
```

```python
n = int(input())
steps = 0

while True:
    s = str(n)
    if s == s[::-1]:     # kiểm tra palindrome
        break

    rev = int(s[::-1])   # đảo số
    n = n + rev          # cộng lại
    steps += 1

print(steps, n)
```

```python
n = int(input())
found = False

i = 1
while i * i <= n:
    sq = i * i
    s = str(sq)

    ok = True
    j = 0
    while j < len(s):
        if s.count(s[j]) > 1:
            ok = False
            break
        j += 1

    if ok:
        print(sq, end=" ")
        found = True

    i += 1

if not found:
    print("(no number)")
```

```python
n = int(input())

best_x = 1
best_len = 1

x = 1
while x <= n:
    cnt = 1
    t = x

    # Tính độ dài dãy Collatz của x
    while t != 1:
        if t % 2 == 0:
            t = t // 2
        else:
            t = 3 * t + 1
        cnt += 1

    # Cập nhật kết quả tốt nhất
    if cnt > best_len or (cnt == best_len and x < best_x):
        best_len = cnt
        best_x = x

    x += 1

print(best_x, best_len)
```

```python
n = int(input())
count_even = 0

i = 1
while i * i <= n:
    if n % i == 0:
        # i là ước
        if i % 2 == 0:
            count_even += 1

        # n // i là ước còn lại (khác i nếu không phải căn bậc 2)
        if i != n // i and (n // i) % 2 == 0:
            count_even += 1
    i += 1

print(count_even)
```

```python
a, b = map(int, input().split())

# Tính tổng ước của a (trừ a)
sum_a = 0
i = 1
while i * i <= a:
    if a % i == 0:
        if i != a:
            sum_a += i
        if i != a // i and a // i != a:
            sum_a += a // i
    i += 1
```
