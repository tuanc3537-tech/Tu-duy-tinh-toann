```python
def binary_search(arr, left, right, target):
    while left <= right:
        mid = (left + right) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    return -1

arr = [1,2,3,4,5,6,7,8,9]
print(binary_search(arr, 0, len(arr)-1, 5))
```

```python
def bublesort():
    n = len(arr)
    for i in range(n-1):
        for j in range(n-1-i):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1]= arr[j+1], arr[j]
    return arr
arr=[1,2,4,2,5,7,8,9]
print(bublesort())

```

```python
def selection_sort(arr):
    n = len(arr)
    for i in range(n):
        minn = i
        for j in range(i+1, n):
            if arr[j] < arr[minn]:
                minn = j
        arr[i], arr[minn] = arr[minn], arr[i]
    return arr

arr = [1,2,3,5,4,7,8,9,10]
print(selection_sort(arr))      
```

```python
def most_frequent(arr):
    count = {}
    best = None
    best_count = 0

    for x in arr:
        count[x] = count.get(x, 0) + 1
        if count[x] > best_count: 
            best = x
            best_count = count[x]

    return best
arr=[1,1,1,2,2,2,2,1]
print(most_frequent(arr))
```

```python
def daytang(nums):
    n= len(nums)
    dp= [1]*n
    for i in range(n):
        for j in range(i):
            if nums[j] < nums[i]:
                dp[i]=max(dp[i], dp[j]+1)
    return max(dp)
nums =[1,2,2,5,4,78,9,6,3,21,5,5,8]
print(daytang(nums))
```

```python
def max_mod(nums, m):
    kq = 0
    
    for i in range(len(nums)):
        tong = 0
        for j in range(i, len(nums)):
            tong += nums[j]
            kq = max(kq, tong % m)

    return kq

nums = [3, 3, 9, 9, 5]
m = 7
print(max_mod(nums, m))
```

```python
def day_hai_hoa_dai_nhat(arr):
    from collections import Counter
    
    tan_suat = Counter(arr)   
    kq = 0
    
    for x in tan_suat:
        if x + 1 in tan_suat:    
            kq = max(kq, tan_suat[x] + tan_suat[x + 1])
    
    return kq


arr = [1, 3, 2, 2, 5, 2, 3, 7]
print(day_hai_hoa_dai_nhat(arr))
```
