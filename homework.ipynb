```python
c = input("")
if c.isalpha():
    print(f"{c} là kí tự alphabet")
else:
    print(f"{c} không phải là kí tự alphabet")
```

```python
c = input("")
if c.isupper():
    if c == 'A':
        print("A")
    else:
        lower_char = c.lower()
        previous_char = chr(ord(lower_char) - 1)
        print(f"Chữ cái thường liền trước là: {previous_char}")
else:
    print("Ký tự nhập vào không phải là chữ hoa.")
```

```python
a = int(input(""))
b = int(input(""))
c = int(input(""))
if a+b>c and b+c>a and a+c>b :
    p = (a+b+c)/2
    s = (p*(p-a)*(p-b)*(p-c))**0.5
    print("dien tich ram giac : {:.1f}".format(s))
else :
    print("k phai tam giac")
```

```python
ten = input("")
cu = int(input())
moi = int(input())
dien = moi-cu
if dien <= 50:
   tien = dien*1984
elif 51 <= dien <= 100:
   tien = dien*2050
elif 101 <= dien <= 200:
   tien = dien*2380
elif 201<= dien <= 300:
   tien = dien*2998
elif 301 <= dien <= 400:
   tien = dien*3350
else:
   tien = dien*3460
tien_phai_tra = round(tien*108/100)
print(ten)
print(tien_phai_tra)
```
