```python
max_val = None
min_val = None

while True:
    x = int(input())
    if x == -1:
        break
    if max_val is None: 
        max_val = x
        min_val = x
    else:
        if x > max_val:
            max_val = x
        if x < min_val:
            min_val = x

print("So lon nhat:", max_val)
print("So nho nhat:", min_val)
```
