```python
def hamxc():
    i= 0
    j= 0
    c =[]
    while i < len(a) and j < len(b):
        if a[i]<b[j]:
            c.append(a[i])
            i+=1
        else:
            c.append(b[j])
            j+=1
    c.extend(a[i:])
    c.extend(b[j:])
    return c
a = list(map(int, input().split()))
b = list(map(int, input().split()))
print(hamxc())
```
