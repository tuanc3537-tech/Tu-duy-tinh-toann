```python
def canPlaceFlowers(flowerbed, k):
    n = len(flowerbed)
    count = 0

    for i in range(n):
        if flowerbed[i] == 0:
            left = (i == 0) or (flowerbed[i-1] == 0)
            right = (i == n-1) or (flowerbed[i+1] == 0)

            if left and right:
                flowerbed[i] = 1   
                count += 1

                if count >= k:
                    return True

    return count >= k
k = int(input())
flowerbed = list(map(int, input().split()))
print(canPlaceFlowers(flowerbed, k))

```

```python
def Median(nums):
    a, b, c, d, e = nums

    if a > b:
        a, b = b, a
    if c > d:
        c, d = d, c
    if a > c:
        a, c = c, a
        b, d = d, b
    if b > e:
        b, e = e, b
    if b > c:
        b, c = c, b
    if c > e:
        c = e

    return c


nums = list(map(int, input().split()))  
print(Median(nums))
```

```python
def findRadius(houses, heaters):
    houses.sort()
    heaters.sort()

    i = 0
    res = 0

    for h in houses:
        # tìm lò sưởi gần nhất bên phải của h
        while i < len(heaters) - 1 and heaters[i] < h:
            i += 1

        if heaters[i] < h:
            dist = h - heaters[i]
        else:
            if i == 0:
                dist = heaters[i] - h
            else:
                dist = min(heaters[i] - h, h - heaters[i - 1])

        res = max(res, dist)

    return res
```
